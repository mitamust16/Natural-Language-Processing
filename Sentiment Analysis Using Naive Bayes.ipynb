## Praproses Teks

```python
# Import data
data = pd.read_excel('D:/VALIDASI/Data TA/Belviu Hotel2.xlsx')
```

```python
x = data['Review'] #get comment
y = data['Sentimen'] #get class
```

```python
y.value_counts()
```

```python
lower = []
for line in x:
    a = line.lower()
    lower.append(a)
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemmed = map(lambda x: stemmer.stem(x), lower)
translator = str.maketrans('', '', string.punctuation)
no_punc = map(lambda x: x.lower().translate(translator),stemmed)
```

```python
stopword = open("E:/stopwords/stopwords indonesia.txt","r").read()
final = []
for line in no_punc:
    word_token = nltk.word_tokenize(line)
    word_token = [word for word in word_token if not word in stopword and not word[0].isdigit()]
    final.append(" ".join(word_token))
```

```python
df_xtrain = pd.DataFrame(final)
df_xtrain.columns=['Review']
df_ytrain = pd.DataFrame(y)
df_xytrain = pd.concat([df_xtrain,df_ytrain],axis=1)
df_xytrain.to_csv('D:/VALIDASI/Praproses/belviu.csv', header=True, index=False, sep=',')
```

## TF-IDF

```python
#TF-IDF
count_vectorizer=CountVectorizer(binary=True)
countvec_x=count_vectorizer.fit_transform(x)
tfidf_x=TfidfTransformer(use_idf=True).fit_transform(countvec_x)
```

## Naive Bayes Classifier (tanpa SMOTE)

```python
accuracy1=[]
accuracy2=[]
AUC1=[]
AUC2=[]
nb = BernoulliNB()
kf=KFold(n_splits=10, shuffle=False)
i=1
for train_index, test_index in kf.split(tfidf_x,y):
    print("Fold ",i)
    print("TRAIN :",train_index,"TEST :",test_index)
    x_train ,x_test=tfidf_x[train_index] ,tfidf_x[test_index]
    y_train, y_test=y[train_index], y[test_index]
    i+=1
    
    classify_train=nb.fit(x_train,y_train)
    #prob_train=classify_train.predict_proba(x_train[:10])
    ypred_trainn=classify_train.predict(x_train)
    
    #prob_test=classify_train.predict_proba(x_test[:10])
    ypred_test=classify_train.predict(x_test)
    
    #Evaluasi
    score1=accuracy_score(y_train,ypred_trainn)
    score2=accuracy_score(y_test,ypred_test)
    accuracy1.append(score1)
    accuracy2.append(score2)
    AUC_score1=roc_auc_score(y_train,ypred_trainn)
    AUC1.append(AUC_score1)
    AUC_score2=roc_auc_score(y_test,ypred_test)
    AUC2.append(AUC_score2)
   # print("Probabilitas Klasifikasi NBC Data Training")
   # print(prob_train)
   # print("Probabilitas Klasifikasi NBC Data Testing")
   # print(prob_test)
    print("CONF MATRIX TRAINING TANPA SMOTE")
    print(confusion_matrix(y_train,ypred_trainn))
    print("CONF MATRIX TESTING TANPA SMOTE")
    print(confusion_matrix(y_test, ypred_test))
    print("=================================================================================")

print("AKURASI TRAINING TANPA SMOTE")
print(accuracy1)
print("AKURASI TESTING TANPA SMOTE")
print(accuracy2)
print("AUC TRAINING TANPA SMOTE")
print(AUC1)
print("AUC TESTING TANPA SMOTE")
print(AUC2)
```

## Rata-rata akurasi dan AUC 10-Fold CV

```python
# SEBELUM SMOTE
print("Sebelum SMOTE:")
jumlah=0
for i in accuracy1:
    jumlah += i
    mean = jumlah / len(accuracy1)
print("Mean accuracy training: ", mean)

jumlah1=0
for j in accuracy2:
    jumlah1 += j
    mean1 = jumlah1 / len(accuracy2)
print("Mean accuracy testing: ", mean1)

jumlah2=0
for k in AUC1:
    jumlah2 += k
    mean2 = jumlah2 / len(AUC1)
print("Mean AUC training: ", mean2)

jumlah3=0
for l in AUC2:
    jumlah3 += l
    mean3 = jumlah3 / len(AUC2)
print("Mean AUC testing: ", mean3)

print("CONF MATRIX TRAINING TANPA SMOTE")
print(confusion_matrix(y_train,ypred_trainn))
print("CONF MATRIX TESTING TANPA SMOTE")
print(confusion_matrix(y_test, ypred_test))
```
